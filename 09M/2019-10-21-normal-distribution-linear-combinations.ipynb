```r
N  <- 5000
Y1 <- rnorm(N, 100, 15)
plot(Y1)
```

```r
plot(density(Y1))
```

```r
Y2 <- 0.4*Y1 + rnorm(N, 0, 10)
plot(Y2)
```

```r
plot(density(Y2))
```

```r
plot(Y1, Y2, pch=16, col=rgb(0, 0, 1, 0.3))
```

```r
smoothScatter(Y1, Y2, bandwidth=4)
```

```r
install.packages("hexbin")
```

```r
library(hexbin)
res <- hexbin(Y1, Y2, xbins=20)
plot(res)
```

```r
plot(Y1 + Y2, pch=16, col=rgb(0, 1, 0, 0.3))
plot(density(Y1+Y2))
```
